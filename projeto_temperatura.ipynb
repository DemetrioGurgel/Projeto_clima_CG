```python
app = Dash(__name__)

# assume you have a "long-form" data frame
# see https://plotly.com/python/px-arguments/ for more options
plotly = df
plotly2 = df2
plotly3 = df3

fig = px.bar(plotly, x="Ano", y="Média max", color="Ano")
fig2 = px.bar(plotly2, x="Ano", y="Média min", color="Ano")
fig3 = px.bar(plotly3, x="Ano", y="Média b. seco", color="Ano")
#histogram

app.layout = html.Div(children=[
    html.H1(children='Temperatura Campina Grande'),

    html.Div(children='''
        Gráficos de médias de temperatura na cidade de Campina Grande entre os anos de 2010 a 2021.
    '''),

    dcc.Graph(
        id='Gráfico de média máxima',
        figure=fig
    ),
    dcc.Graph(
        id='Gráfico de média mínima',
        figure=fig2
    ),
    dcc.Graph(
        id='Gráfico de média no bulbo seco (desconsiderando a umidade)',
        figure=fig3
    )
])

if __name__ == '__main__':
    app.run_server(debug=False)
```
